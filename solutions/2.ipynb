```python
from pathlib import Path
```

```python
transcript = Path('../data/shrek2.txt').read_text()
```

```python
lines = transcript.split('\n')
```

```python
shrek_lines = [line for line in lines if line.startswith('Shrek:')]
```

```python
shrek_lines
```

```python
donkey_count = 0
```

```python
for shrek_line in shrek_lines:
    donkey_count += shrek_line.lower().count('donkey')
```
