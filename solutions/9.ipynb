```python
from pathlib import Path
```

```python
transcript = Path('../data/shrek2.txt').read_text()
```

```python
CHARS_TO_FIND = 'bduzoq otmdyuzs'
```

```python
lowercase_alphabet = list('abcdefghijklmnopqrstuvwxyz')
```

```python
def caesar_cipher( string: str, key: int ) -> str:

    encryped_string = ''
    for char in string:
        if char.lower() in lowercase_alphabet:
            char = char.lower()
            char_index = lowercase_alphabet.index(char)
            new_index = (char_index + key) % len(lowercase_alphabet)
            encryped_string += lowercase_alphabet[new_index]

        else:
            encryped_string += char

    return encryped_string
```

```python
caesar_cipher('abcd?', 1)
```

```python
encrypted_transcript = caesar_cipher(transcript, 12)
```
